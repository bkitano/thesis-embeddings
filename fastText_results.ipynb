```python
import torch
import pandas as pd
```

```python
emb = pd.read_csv('./vard_dwe/model.vec', 
                  sep=' ', 
                  skiprows=[0], 
                  header=None) 
```

```python
# set the column names to be the word, and create the PPMI id column
emb.columns = ['word', *list(range(100)), 'ID']
```

```python
emb
```

```python
from gensim.corpora import Dictionary
dct = Dictionary.load('../corpus/vard.dct')
```

```python
words = [p[1] for p in dct.items()]
```

```python
# find which words are in ppmi (rectified embeddings)
rectEmb = emb[emb.word.isin(words)]
```

```python
rectEmb.shape
```

```python
# replace the ID col vals with the ppmi id
rectEmb['ID'] = rectEmb['word'].map(lambda w: dct.token2id[w])
```

```python
# sort, since the most freqent words have lower ppmi id's
rectEmb = rectEmb.sort_values(by=['ID'])
```

```python
rectEmb
```

```python
# create new column to store the embedding id's
rectEmb['embID'] = list(range(rectEmb.shape[0]))
```

```python
rectEmb
```

```python
import pickle
# create ppmi to emb mappings
dicter = rectEmb.values[:,[0,-2,-1]]
ppmiIDToEmbID = dict(zip(dicter[:,-2].tolist(), dicter[:,-1].tolist()))
EmbIDToPPMIId = dict(zip(dicter[:,-1].tolist(), dicter[:,-2].tolist()))

with open('./vard_ppmiIDToEmbID.p', 'wb') as fp:
    pickle.dump(ppmiIDToEmbID, fp)
    
with open('./vard_embIDToPPMIID.p', 'wb') as fp:
    pickle.dump(EmbIDToPPMIId, fp)
```

```python
# now create an embedding matrix that torch can ingest
embeddings = rectEmb.values
embeddings[:,0] = embeddings[:,-1] # replace the word col with the emb ID
embeddings = embeddings[:,:-2] # remove the last two id cols
```

```python
embeddings.shape
```

```python
embeddings
```

```python
with open('./vard_staticEmbeddings.p', 'wb') as fp:
    pickle.dump(embeddings, fp)
```
